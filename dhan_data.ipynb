```python
async def main():
    await fd.connect()

    bid_cache = {}

    while True:
        raw = await fd.ws.recv()
        remaining = raw

        while remaining:
            update = fd.process_data(remaining)
            if not update:
                break

            remaining = update.pop("remaining_data", None)
            sid = update["security_id"]

            if update["type"] == "Bid":
                bid_cache[sid] = update

            elif update["type"] == "Ask" and sid in bid_cache:
                formatted = fd.combine_and_format_depth(
                    bid_cache[sid],
                    update
                )

                print(f"\nSecurity ID: {sid}")
                for row in formatted["depth"]:
                    print(row)

                del bid_cache[sid]

await main()
```
